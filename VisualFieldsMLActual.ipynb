```python
# Get all patients with 4 or more fields,
# counting each eye as a different patient to offset loss of data

final_patients = {}

for i in range(len(df)):
   if (i + 1) < len(df):
    if (df.loc[i, "Eye"] != df.loc[i+1, "Eye"]) and (df.loc[i,"FieldN"] >= 7):
      final_patients[str(df.loc[i,"PatID"])+'_'+str(df.loc[i, "Eye"])] = i - df.loc[i,"FieldN"] + 1
```

```python
def flat_to_image(vector):
  # 4 6 8 10 10 8 6 4
  nums = [4,6,8,9,9,8,6,4]
  pads = [[3,2],[2,1],[1,0],[0,0],[0,0],[1,0],[2,1],[3,2]]
  ret = []
  idx=0
  for i in range(len(nums)):
    row = []
    for j in range(nums[i]):
      row.append(vector[idx])
      idx += 1
    for k in range(pads[i][0]):
      row.insert(0,100)
    for k in range(pads[i][1]):
      row.append(100)
    ret.append(row)
  return ret
```

```python
def data_prep_masking():
  x = []
  y = []
  for index in final_patients.values():
    x_entry = []
    y_entry = []
    correct_id = df.loc[index, "PatID"]
    correct_eye = df.loc[index, "Eye"]
    for i in range(8):
      curr_index = index + i
      if df.loc[curr_index, "PatID"] == correct_id and df.loc[curr_index, "Eye"] == correct_eye:
        x_test = []
        y_test = []
        for col in list(df.columns):
          if "Sens" in col:
            x_test.append(df.loc[curr_index, col])
            y_test.append(df.loc[curr_index, col])
        x_entry.append(flat_to_image(x_test))
        y_entry.append(flat_to_image(y_test))
      else:
        x_entry.append(np.full((8,9), -1))
    y.append(y_entry[-1])
    del x_entry[-1]
    x.append(x_entry)
  return x,y

x,y = data_prep_masking()
```

```python
def data_prep_6inputs():
  x = []
  y = []
  for index in final_patients.values():
    x_entry = []
    y_entry = []
    correct_id = df.loc[index, "PatID"]
    correct_eye = df.loc[index, "Eye"]
    for i in range(7):
      curr_index = index + i
      x_test = []
      y_test = []
      for col in list(df.columns):
        if "Sens" in col:
          x_test.append(df.loc[curr_index, col])
          y_test.append(df.loc[curr_index, col])
      x_entry.append(flat_to_image(x_test))
      y_entry.append(flat_to_image(y_test))
    y.append(y_entry[-1])
    del x_entry[-1]
    x.append(x_entry)
  return x,y

x,y = data_prep_6inputs()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

np.random.seed(42)
x, y = shuffle(x, y)

x_train, x_testing, y_train, y_testing = train_test_split(x, y, test_size = .1)
```

```python
x_testing.shape
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, LSTM, Flatten, Dense, MaxPooling2D, TimeDistributed, Reshape, Dropout, Masking, Input, ConvLSTM2D, BatchNormalization
from keras.layers import ConvLSTM2D
from tensorflow import keras

inp = Input(shape=(6,8,9,1))

# We will construct 3 `ConvLSTM2D` layers with batch normalization,
# followed by a `Conv3D` layer for the spatiotemporal outputs.
#x = Masking(mask_value = -.01)(inp)
x = ConvLSTM2D(
    filters=64,
    kernel_size=(5, 5),
    padding="same",
    return_sequences=True,
    activation="relu",
)(inp)
x = BatchNormalization()(x)
x = ConvLSTM2D(
    filters=64,
    kernel_size=(3, 3),
    padding="same",
    return_sequences=True,
    activation="relu",
)(x)
x = BatchNormalization()(x)
x = ConvLSTM2D(
    filters=64,
    kernel_size=(1, 1),
    padding="same",
    return_sequences=True,
    activation="relu",
)(x)
x = Flatten()(x)
x = Dense(72, activation='linear')(x)
x = Reshape((8,9,1))(x)

# Next, we will build the complete model and compile it.
model = keras.models.Model(inp, x)
model.compile(
    loss="mse",
    optimizer=keras.optimizers.Adam(),
)

early_stopping = keras.callbacks.EarlyStopping(monitor="val_loss", patience=10)
reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor="val_loss", patience=5)

history = model.fit(x_train, y_train, epochs=100, callbacks=[early_stopping, reduce_lr], validation_split=.1)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model train vs validation loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, LSTM, Flatten, Dense, MaxPooling2D, TimeDistributed, Reshape

model = Sequential()
model.add(TimeDistributed(Conv2D(64, (3,3), activation="relu", input_shape = (None,4,8,9,1))))
model.add(TimeDistributed(MaxPooling2D(pool_size=(2,2),strides=(1,1))))
model.add(TimeDistributed(Conv2D(64, (2,2), activation='relu')))
model.add(TimeDistributed(MaxPooling2D(pool_size=(2,2),strides=(1,1))))
model.add(TimeDistributed(Flatten()))
model.add(LSTM(100, activation='relu'))
model.add(Dense(72))
model.add(Reshape((8,9,1)))

model.compile(optimizer="adam", loss="mse")

early_stopping = keras.callbacks.EarlyStopping(monitor="val_loss", patience=10)
reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor="val_loss", patience=5)

history = model.fit(x_train, y_train, epochs=500, verbose=2, callbacks=[early_stopping, reduce_lr], validation_split=.1)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model train vs validation loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```

```python
from keras.layers import Input, Dense, Conv2D, Flatten, concatenate, LeakyReLU, Activation
from keras.initializers import RandomNormal
from keras.layers import BatchNormalization, Conv2DTranspose, Dropout, Concatenate
from keras.optimizers import Adam
from keras.models import Model
import tensorflow as tf
from tensorflow import keras
from matplotlib import pyplot as plt
import numpy as np

def define_discriminator():
	# weight initialization
	init = RandomNormal(stddev=0.02)
	# source image input
	in_var = Input(shape = (2))
	v = Dense(4, activation = 'relu')(in_var)
	v = Dense (8, activation = 'relu')(v)
	v = Dense (16, activation = 'relu')(v)
	v = Dense (16, activation = 'relu')(v)
  #images input path
	in_src = Input(shape = (8,9,1))
	in_trgt = Input(shape = (8,9,1))
	# concatenate images channel-wise
	merged = Concatenate()([in_src, in_trgt])
	# C64
	d = Conv2D(64, (4,4), strides=(2,2), padding='same', kernel_initializer=init)(merged)
	d = LeakyReLU(alpha=0.2)(d)
	# C128
	d = Conv2D(128, (4,4), strides=(2,2), padding='same', kernel_initializer=init)(d)
	d = BatchNormalization()(d)
	d = LeakyReLU(alpha=0.2)(d)
	# C256
	d = Conv2D(256, (4,4), strides=(2,2), padding='same', kernel_initializer=init)(d)
	d = BatchNormalization()(d)
	d = LeakyReLU(alpha=0.2)(d)
	# C512
	d = Conv2D(512, (4,4), strides=(2,2), padding='same', kernel_initializer=init)(d)
	d = BatchNormalization()(d)
	d = LeakyReLU(alpha=0.2)(d)
	# second last output layer
	d = Conv2D(512, (4,4), padding='same', kernel_initializer=init)(d)
	d = BatchNormalization()(d)
	d = LeakyReLU(alpha=0.2)(d)
	# patch output
	d = Flatten()(d)
	d = concatenate([d, v])
	d = Dense(128, activation = 'relu')(d)
	d = Dense(128, activation = 'relu')(d)
	d = Dense(64, activation = 'relu')(d)
	d = Dense(1)(d)
	d = Reshape((1,1,1))(d)
	d = Conv2D(1, (4,4), padding='same', kernel_initializer=init)(d)
	patch_out = Activation('sigmoid')(d)

	# define model
	model = Model([in_src, in_trgt, in_var], d)
	# compile model
	opt = Adam(lr=0.0002, beta_1=0.5)
	model.compile(loss='binary_crossentropy', optimizer=opt, loss_weights=[0.5])
	model.summary()
	return model
```

```python
from keras.layers import MaxPool2D, UpSampling2D, Reshape, ZeroPadding2D, Cropping2D, Flatten, concatenate, Dense
def create_autoencoder(ls_size):
  in_src = Input(shape = (8,9,1))

  in_var = Input(shape=(2))

  x = in_src
  # down path
  start_filters = 4
  for i in range(2):
      x = Conv2D(start_filters * (2**i), 3, padding='same', activation='relu')(x)
      x = Conv2D(start_filters * (2**i), 3, padding='same', activation='relu')(x)
      x = MaxPool2D()(x)
  # bottleneck
  x = Conv2D(start_filters*(2**ls_size),3, padding='same', activation='relu')(x)
  x = Flatten()(x)
  x = concatenate([x, in_var])
  x = Dense(128, activation = 'relu')(x)
  x = Reshape((2,2,32))(x)
  x = Conv2D(start_filters*(2**ls_size),3, padding='same', activation='relu')(x)
  # up path
  for i in range(2):
      x = UpSampling2D((3,3))(x)
      x = Conv2D(start_filters*(2**(2-i)),3, padding='same', activation='relu')(x)
      x = Conv2D(start_filters*(2**(2-i)),3, padding='same', activation='relu')(x)
  output = Conv2D(1, 1, activation='sigmoid')(x)
  cropped = Cropping2D(cropping=((10,0),(9,0)))(output)
  model = tf.keras.Model([in_src, in_var], cropped)
  model.summary()
  return model

create_autoencoder(3)
```

```python
def define_gan(g_model, d_model):
 # make weights in the discriminator not trainable
 for layer in d_model.layers:
  if not isinstance(layer, BatchNormalization):
    layer.trainable = False
 # define the source image
 in_src = Input(shape=(8,9,1))
 in_var = Input(shape = (2))
 # connect the source image to the generator input
 gen_out = g_model([in_src, in_var])
 # connect the source input and generator output to the discriminator input
 dis_out = d_model([in_src, gen_out, in_var])
 # src image as input, generated image and classification output
 model = Model([in_src, in_var], [dis_out, gen_out], name = 'gan')
 # compile model
 opt = Adam(learning_rate=0.0002, beta_1=0.5)
 model.compile(loss=['binary_crossentropy', 'mae'], optimizer=opt, loss_weights=[1,100])
 model.summary()
 return model
```

```python
from matplotlib import pyplot as plt
import pickle
#x1 => source
#x2 => vars
#x3 => target

data = [x1_test, x2_test, x3_test]
def generate_real_samples(dataset, n_samples, patch_shape):
  src, var, trgt = dataset
  ix=np.random.randint(0, src.shape[0], n_samples)
  x1,x2,x3 = src[ix], var[ix], trgt[ix]
  y = np.ones((n_samples, patch_shape, patch_shape))
  return [x1,x2,x3], y

def generate_fake_samples(gen, samples, vars, patch_shape):
  x = gen.predict([samples, vars])
  y = np.zeros((len(x), patch_shape, patch_shape, 1))
  return x, y


# generate samples and save as a plot and save the model
def summarize_performance(step, g_model, dataset, n_samples=3):
  # select a sample of input images
  [X_realA, X_vars, X_realB], _ = generate_real_samples(dataset, n_samples, 1)
  # generate a batch of fake samples
  X_fakeB, _ = generate_fake_samples(g_model, X_realA, X_vars, 1)

  save = {'X_vars': X_vars, 'X_realA': X_realA, 'X_realB': X_realB, 'X_fakeB': X_fakeB}
  filename1 = 'plot_%06d.pickle' % (step+1)


  filehandler = open(filename1,"wb")
  pickle.dump(save,filehandler)
  filehandler.close()

  # save the generator model
  filename2 = 'model_%06d.h5' % (step+1)
  g_model.save(filename2)
  print('>Saved: %s and %s' % (filename1, filename2))
```

```python
# train pix2pix model
def train(d_model, g_model, gan_model, dataset, n_epochs=50, n_batch=1):
 # determine the output square shape of the discriminator
 n_patch = d_model.output_shape[1]
 # unpack dataset
 trainA, vars, trainB = dataset
 # calculate the number of batches per training epoch
 bat_per_epo = int(len(trainA) / n_batch)
 # calculate the number of training iterations
 n_steps = bat_per_epo * n_epochs
 print(n_steps)
 # manually enumerate epochs
 for i in range(n_steps):
  # select a batch of real samples
  [X_realA, X_vars, X_realB], y_real = generate_real_samples(dataset, n_batch, n_patch)
  # generate a batch of fake samples
  X_fakeB, y_fake = generate_fake_samples(g_model, X_realA, X_vars, n_patch)
  # update discriminator for real samples
  d_loss1 = d_model.train_on_batch([X_realA, X_realB, X_vars], y_real)
  # update discriminator for generated samples
  d_loss2 = d_model.train_on_batch([X_realA, X_fakeB, X_vars], y_fake)
  # update the generator
  g_loss, _, _ = gan_model.train_on_batch([X_realA, X_vars], [y_real, X_realB])
  # summarize performance
  print('>%d, d1[%.3f] d2[%.3f] g[%.3f]' % (i+1, d_loss1, d_loss2, g_loss))
  # summarize model performance
  epoch = 1
  if (i+1) % (bat_per_epo * 10) == 0:
    summarize_performance(i, g_model, dataset)
    print('completed epoch ', epoch)
    epoch +=1
```

```python
def flat_to_image_pad(vector):
  # 4 6 8 10 10 8 6 4
  nums = [4,6,8,9,9,8,6,4]
  pads = [[3,2],[2,1],[1,0],[0,0],[0,0],[1,0],[2,1],[3,2]]
  ret = []
  idx=0
  for i in range(len(nums)):
    row = []
    for j in range(nums[i]):
      row.append(vector[idx])
      idx += 1
    for k in range(pads[i][0]):
      row.insert(0,100)
    for k in range(pads[i][1]):
      row.append(100)
    ret.append(row)
  return ret
```
